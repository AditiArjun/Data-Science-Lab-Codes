```python
import pandas as pd 
import numpy as np
import statistics as stat
import matplotlib.pyplot as plt 
```

```python
df=pd.read_csv('Iris.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
summary_stats = df.groupby('SepalLengthCm')['SepalWidthCm'].describe()
print(summary_stats)
```

```python
summary_stats = df.groupby('PetalLengthCm')['PetalWidthCm'].describe()
print(summary_stats)
```

```python

df.groupby('Species')[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']].mean()
```

```python

df.groupby('Species')[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']].median()
```

```python

df.groupby('Species')[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']].min()
```

```python

df.groupby('Species')[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']].max()
```

```python

df.groupby('Species')[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']].quantile()
```

```python
df.groupby('Species')[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']].quantile(0.25)
```
